# 2 SVM

```python
import numpy as np
import linear_model_and_liner_classifier
np.random.seed(42)
for c in [0.5, 1, 10, 100]:
    linear_model_and_liner_classifier.run_svm(c)
```

**可以看对于线性可分数据集，c的大小并不影响最终结果（不影响决策平面与边界平面的斜率、截距），对于线性可分数据集是没有c的，或者认为c为+∞**
**对于线性不可分数据，可以看出当C变化的确会影响错分样本数目（c=0.5错分5个，c=100错分2个）**
